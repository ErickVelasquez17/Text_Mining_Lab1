## **Problema 1**

```
def problema1 (imagen, color):
  import cv2
  import matplotlib.pyplot as plt
  import numpy as np

  img = cv2.imread(imagen)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

  alto = img.shape[0]
  ancho = img.shape[1]

  lienzo = np.zeros((alto, ancho, 3))

  #Ciclo for para definir los colores
  for i in range(0, alto):
    for j in range(0, ancho):
        pixel = img[i, j]
        
        blue = pixel[2]
        green = pixel[1]
        red = pixel[0]
        
        #Condiciones IF para seleccionar el color seg√∫n el parametro ingresado.
        #Se escribe en BGR
        if color == 1:                    
          lienzo[i, j] = [blue, 0, 0]
        elif color == 2:
          lienzo[i, j] = [0, green, 0]
        elif color == 3:
          lienzo[i, j] = [0, 0, red]
        elif color == 10:                    
          lienzo[i, j] = [0, green, red]
        elif color == 20:
          lienzo[i, j] = [blue, green, 0]
        elif color == 30:
          lienzo[i, j] = [blue, 0, red]

  #Se baja la imagen a disco
  cv2.imwrite("monocromo.jpg", lienzo)
  monocromo = cv2.imread("monocromo.jpg")
  monocromo = cv2.cvtColor(monocromo, cv2.COLOR_BGR2RGB) 

  plt.imshow(monocromo)
  plt.show()    
```

# **Problema 3**

```
def problema3 (imagen):
  import cv2
  import matplotlib.pyplot as plt
  import numpy as np

  img = cv2.imread(imagen)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

  alto = img.shape[0]
  ancho = img.shape[1]

  media_ponderada_blue = np.zeros((alto, ancho, 1)) #Escala de grises
  media_ponderada_red = np.zeros((alto, ancho, 1)) #Escala de grises
  media_ponderada_green = np.zeros((alto, ancho, 1)) #Escala de grises

  #Ciclo for para definir los colores
  for i in range(0, alto):
    for j in range(0, ancho):
        pixel = img[i, j]
        
        blue = pixel[2]
        green = pixel[1]
        red = pixel[0]
        
        #Condiciones IF para seleccionar el color seg√∫n el parametro ingresado.
        #Se escribe en BGR

        media_ponderada_blue[i, j] = int(blue)
        media_ponderada_green[i, j] = int(green)
        media_ponderada_red[i, j] = int(red)

  #Se baja la imagen a disco
  cv2.imwrite("gris_ponderado_blue.jpg", media_ponderada_blue)
  gris_blue = cv2.imread("gris_ponderado_blue.jpg")
  gris_blue = cv2.cvtColor(gris_blue, cv2.COLOR_BGR2RGB)

  cv2.imwrite("gris_ponderado_green.jpg", media_ponderada_green)
  gris_green = cv2.imread("gris_ponderado_green.jpg")
  gris_green = cv2.cvtColor(gris_green, cv2.COLOR_BGR2RGB)

  cv2.imwrite("gris_ponderado_red.jpg", media_ponderada_red)
  gris_red = cv2.imread("gris_ponderado_red.jpg")
  gris_red = cv2.cvtColor(gris_red, cv2.COLOR_BGR2RGB)

  print("Gris Azul:")
  plt.imshow(gris_blue)
  plt.show()

  print("Gris Verde:")
  plt.imshow(gris_green)
  plt.show()
  
  print("Gris Rojo:")
  plt.imshow(gris_red)
  plt.show()
```

# **Problema 4**

```
def problema4(imagen):
  import cv2
  import numpy as np
  from matplotlib import pyplot as plt
  
  img = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Text Mining/Estadio.jpeg")
  color = ('b','g','r')

  #Histograma Rojo:
  for i,col in enumerate(color):
      histr_red = cv2.calcHist([img],[0],None,[256],[0,256])
      plt.plot(histr_red,color = 'red')
      plt.xlim([0,256])

  plt.show()

  #Histograma verde:
  for i,col in enumerate(color):
      histr_green = cv2.calcHist([img],[1],None,[256],[0,256])
      plt.plot(histr_green,color = 'green')
      plt.xlim([0,256])

  plt.show()

  #Histograma azul:
  for i,col in enumerate(color):
      histr_blue = cv2.calcHist([img],[2],None,[256],[0,256])
      plt.plot(histr_blue,color = 'blue')
      plt.xlim([0,256])

  plt.show()

  #Histograma gris:
  img = cv2.imread(imagen)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

  alto = img.shape[0]
  ancho = img.shape[1]

  media_ponderada = np.zeros((alto, ancho, 1)) #Escala de grises

  #Ciclo for para definir los colores
  for i in range(0, alto):
    for j in range(0, ancho):
        pixel = img[i, j]
        
        blue = pixel[2]
        green = pixel[1]
        red = pixel[0]
        
        #Condiciones IF para seleccionar el color seg√∫n el parametro ingresado.
        #Se escribe en BGR

        media_ponderada[i, j] = int(0.299*blue + 0.587*green + 0.114*red)
        
  #Se baja la imagen a disco
  cv2.imwrite("gris_ponderado.jpg", media_ponderada)
  gris = cv2.imread("gris_ponderado.jpg")
  gris = cv2.cvtColor(gris, cv2.COLOR_BGR2RGB)

  for i,col in enumerate(color):
      histr_gray = cv2.calcHist([gris],[2],None,[256],[0,256])
      plt.plot(histr_gray,color = 'gray')
      plt.xlim([0,256])

  plt.show()
```

Investigue brevemente en que consiste el espacio de color HSV y como se mapean colores a dicho espacio.

El espacio HSV (o HSB) en una representaci√≥n tridimensional del color basada en tres factores, que son:

*   Matiz es lo que com√∫n -y err√≥neamente- llamamos color: la longitud de onda dominante
*   Saturaci√≥n mide la pureza de un color, es decir la ausencia de otros colores: un color con el 100% de saturaci√≥n se conseguir√° tan solo con su matiz monocrom√°tico o espectral.
*   Valor o Brillo hace referencia a la intensidad de la luz, es decir, la cantidad de luz que hay en ese color, desde un 0% (negro) al 100%.

El color HSV sigue una representaci√≥n m√°s parecida a las coordenadas cil√≠ndricas. Adem√°s es una representaci√≥n m√°s cercana a la forma en que los humanos perciben los colores y sus propiedades, pues se agrupan las tonalidades de color, lo cual es distinto al caso RGB donde los colores no est√°n necesariamente tan agrupados.
